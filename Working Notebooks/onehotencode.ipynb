```python
# Split dataset
from sklearn.model_selection import TimeSeriesSplit
from matplotlib import pyplot

tss = TimeSeriesSplit(n_splits = 3)
X = dataset.drop(labels=['Sales'], axis=1)
y = dataset.loc[:,'Sales':'Sales']

for train_index, test_index in tss.split(X):
    X_train, X_test = X.iloc[train_index, :], X.iloc[test_index,:]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
from numpy import array
from numpy import argmax
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
# define example
data = dataset['DayOfWeek']
values = array(data)
print(values)
# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(values)
print(integer_encoded)
# binary encode
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)
# invert first example
inverted = label_encoder.inverse_transform([argmax(onehot_encoded[0, :])])
print(inverted)
```

```python
# Encoding the Independent Variable
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
onehotencoder = OneHotEncoder(categorical_features = [0])
dataset = onehotencoder.fit_transform(dataset).toarray()
```

```python
features = pd.get_dummies(dataset)
features.head(5)
```

```python
dayofweek_one_hot = pd.get_dummies(dataset['DayOfWeek'])
dayofweek_one_hot = dayofweek_one_hot.rename(columns={1: 'Mon', 2: 'Tue',\
                                                 3: 'Wed', 4: 'Thu', 5:'Fri', 6:'Sat', 7:'Sun'}).reset_index(drop=True)

month_one_hot = pd.get_dummies(dataset['month'])
month_one_hot = month_one_hot.rename(columns={1: 'Jan', 2: 'Feb',\
                                                 3: 'Mar', 4: 'Apr', 5:'Mai', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}).reset_index(drop=True)

dataset2 = pd.concat([dataset, month_one_hot, dayofweek_one_hot], axis=1)
dataset2.head()
```

```python
dataset2 = dataset2.drop(columns=['DayOfWeek', 'month'])
dataset2.head()
```
